# Performing over sampling

```python
from imblearn.under_sampling import RandomUnderSampler 
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
```

```python
x=finalDataset.drop(['attrition','award count_x','budget utilization percentage','fresher flag','index',],axis=1)
y=finalDataset['attrition']
```

```python
 #Apply the random under-sampling 
rus = RandomUnderSampler()
X_RUS, y_RUS = rus.fit_resample(x, y)
```

```python
from sklearn.linear_model import LogisticRegression
lr1 = LogisticRegression(solver='liblinear')
lr1.fit(X_RUS, y_RUS)
```

```python
list1 = lr1.coef_.tolist()
len(list1[0])
```

```python
weight4=pd.DataFrame({'varibles':X_RUS.columns,'coeff':list1[0]})
weight4.sort_values(by='coeff',inplace=True)
```
